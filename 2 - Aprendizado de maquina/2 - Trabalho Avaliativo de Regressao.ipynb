#### CARREGANDO OS DADOS

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()

import pandas as pd
dataset = pd.DataFrame(housing.data)
dataset.columns = housing.feature_names
dataset['Price'] = housing.target
dataset.head()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

correlation_matrix = dataset.corr()
plt.rc('font', size=12)
plt.figure(figsize=(8, 7))
sns.heatmap(correlation_matrix, cmap=sns.color_palette("coolwarm"), linewidths=.5, annot=True, fmt='.2f');
plt.title('Matriz de Correla√ß√£o \n')
plt.show()
```

```python
X = dataset['MedInc'].to_frame()
y = dataset['Price'] 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

##### TESTANDO COM O SKLEARN

Utilizem os valores de a (b1), b (b2) e MSE obtidos com o scikit-learn para comparar com os valores obtidos pelo modelo criado por voc√™s.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('a = ' + str(model.coef_))
print('b = ' + str(model.intercept_))

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

plt.figure()
plt.xlabel('Valores Reais')
plt.ylabel('Valores Preditos')
plt.title('Valores Reais vs Valores Preditos')
plt.scatter(y_test, y_pred, label='Valores Preditos')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='-', linewidth=3, label='Linha de Regress√£o')
plt.legend()
plt.grid()
plt.show()
```

#### FA√áA SEU C√ìDIGO AQUI:

```python
#Treinamento 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
#Convertendo X_train e y_train para numpy arrays
X_train = X_train.to_numpy().flatten()
y_train = y_train.to_numpy()
#Calcular coefecientes manualmente
n = len(X_train)
#Somas necess√°rias
sum_x = np.sum(X_train)
sum_y = np.sum(y_train)
sum_x_y = np.sum(X_train * y_train)
sum_x_squared = np.sum(X_train ** 2)
```

```python
#C√°lculo dos coeficientes
beta_1 = (n * sum_x_y - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
beta_0 = (sum_y - beta_1 * sum_x) / n
#Previs√µes
y_pred = beta_0 + beta_1 * X_test.to_numpy().flatten()
print(f'Coeficiente de Inclina√ß√£o (Beta_1): {beta_1}')
print(f'Intercepto (Beta_0): {beta_0}')
```

```python
#Visualizar os resultados
plt.figure(figsize=(10, 6))
#Plotar dados de treinamento e linha de regress√£o
plt.scatter(X_train, y_train, color='blue', label='Dados de Treinamento')
plt.plot(X_train, beta_0 + beta_1 * X_train, color='red', linewidth=2, label='Linha de Regress√£o')
#Plotar valores reais vs. valores preditos
plt.scatter(y_test, y_pred, color='green', alpha=0.5, label='Valores Preditos')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='black', linestyle='--', linewidth=2, label='Linha Ideal')
plt.xlabel('MedInc')
plt.ylabel('Price')
plt.title('Regress√£o Linear Simples: MedInc vs Price')
plt.legend()
plt.grid(True)
plt.show()
```
